```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('SalaryData.csv', usecols = ['Years of Experience','Salary'])
data
```

```python
data = data.dropna()
data
```

```python
x = data.iloc[:,0:1].values # 0 index year of exp -1 vaneko salary but salary lai ligdaina 0 t0 -1 vaneko 0 matra aauxa
x
```

```python
y = data.iloc[:,1:].values
y
```

```python
from sklearn.model_selection import train_test_split
#random state is just to suffle it doesnot matter what you take
#train size min 2/3 i.e min 60%
#test size max 40%
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1,random_state = 10)
```

```python
# print(x_train)
# print(x_test)
# print(y_train)

# print(y_test)
```

```python
print(len(x))
print(len(x_train), len(y_train))
print(len(x_test), len(y_test))
```

```python
#training model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
exp =float(input("enter your years of experience"))
exp = np.array([[exp]])
print(exp)

#Salary predction
predction = model.predict(exp)[0][0]
print(f"Your predicted salary = {predction}")
```

```python
predicated_salary = model.predict(x_test)
for i in range(len(x_test)):
    print(x_test.tolist()[i][0],y_test.tolist()[i][0],predicated_salary[i][0])
```

```python
for i in range(len(x_test)):
    print(f"Years of experience = {x_test.tolist()[i][0]} Actual salary = {y_test.tolist()[i][0]} predicted salary = {predicated_salary[i][0]}")
```
